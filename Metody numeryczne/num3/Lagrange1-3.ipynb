```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import math
import sys
import numpy as np
from horner import horner


f1 = lambda x: horner([1, 0, 3, 0,-1,-1],6, x)
f2 = lambda x: 7.0 * np.cos(x-1) - 2.0 * np.sin(x)
f3 = lambda x: 6**x - 2 * x - 5
f4 = lambda x: 2.0 * np.cos(x) - 1.0**x - x**3.0
f5 = lambda x: np.exp(1/2*x) - 5.0 * np.cos(2*(x-1))
f6 = lambda x: np.absolute(x)
f7 = lambda x: np.absolute(2*np.cos(2*x)+2)


def wykres ():
    fontP = FontProperties()
    fontP.set_size('large')
    p1, = plt.plot(fx, fy, color='deeppink', label='funkcja')
    p2, = plt.plot(ix, iy, color='royalblue', label='Czybyszew')
    p3, = plt.plot(ix2, iy2, color='lime', label='równoodległe')
    for i in range (len(wx)):
        plt.scatter(wx[i], wy[i], color='blue')
    for i in range (len(wx)):
        plt.scatter(wx2[i], wy2[i], color='green')
    plt.grid(True)
    plt.xlim(first,last)
    plt.legend(handles=[p1, p2, p3], bbox_to_anchor=(1.05, 1), loc='upper left', prop=fontP)



def licz_funkcje(funkc, poczatek, koniec, precyzja):
    for i in range (len(fx)):
        fx[i] = poczatek
        fy[i] = funkc(poczatek)
        poczatek = poczatek + precyzja


def licz_wezly(funkc, poczatek, koniec, wezly):
    for i in range (wezly):
        arg = 0
        arg = float(arg)
    
        arg = np.cos(np.pi * (2 * i + 1) / (2 * wezly + 1) )
        arg = ( ( (koniec - poczatek) * arg) + (poczatek + koniec) ) * 0.5
        
        wx[i] = arg
        wy[i] = funkc(arg)
        
def licz_wezly2(funkc, poczatek, koniec, wezly):
    dlugosc = 0
    dlugosc = float(dlugosc)
    step = 0
    step = float(step)
    
    if poczatek * koniec < 0:
        dlugosc = np.absolute(poczatek) + koniec
    else:
        dlugosc = dlugosc = np.absolute(koniec - poczatek)

    step = dlugosc/(wezly-1)
    wx2[0] = poczatek
    wy2[0] = funkc(poczatek)
    wx2[wezly-1] = koniec
    wy2[wezly-1] = funkc(koniec)
    i = 1
    while i < wezly - 1:
        wx2[i] = wx2[i-1] + step
        wy2[i] = funkc(wx2[i])
        i = i + 1


def licz_interp(funkc, poczatek, koniec, wezly):
    i = poczatek
    i = float(i)
    counter = 0
    while counter < len(ix):
        y=0
        y = float(y)
        y2=0
        y2 = float(y)

        for j in range (wezly):
            pomocnicza = wy[j]
            pomocnicza2 = wy2[j]
            pomocnicza=float(pomocnicza)
            pomocnicza2=float(pomocnicza2)
            for k in range (wezly):
                if j!=k:
                    pomocnicza = pomocnicza * ( (i-wx[k]) / (wx[j]-wx[k]))
                    pomocnicza2 = pomocnicza2 * ( (i-wx2[k]) / (wx2[j]-wx2[k]))
            y = y + pomocnicza
            y2 = y2 + pomocnicza2
        
        ix[counter] = i
        iy[counter] = y
        ix2[counter] = i
        iy2[counter] = y2
        counter = counter + 1
        i = i + precision

def sredniBlad():
    sbc = 0
    sbc = float(sbc)
    sbr = 0
    sbr = float(sbr)

    for i in range (len(ix)):
        sbc = sbc + np.absolute(fy[i] - iy[i])
        sbr = sbr + np.absolute(fy[i] - iy2[i])
    sbc = sbc/(len(ix))
    sbr = sbr/(len(ix))
    print('Sredni blad dla wezlow Czebyszewa: ',sbc)
    print('Sredni blad dla wezlow rownoodleglych: ',sbr)

def dostepne_funkcje ():
    print('Dostepne funkcje: ')
    print('[1]  x**5 + 3 * x**3 - x - 1')
    print('[2]  7 * cos(x-1) - 2 * sin(x)')
    print('[3]  6**x - 2 * x - 5')
    print('[4]  2 * cos(x) - 1**x - x**3')
    print('[5]  exp(1/2*x) - 5 * cos(2*(x-1))')
    print('[6]  |x|')
    print('[7]  |2 * cos(2*x) + 2|')
```
