```python
import matplotlib.pyplot as plt
import sys
import numpy as np
import metodaSiecznych
import regulaFalsi
from horner import horner



f1 = lambda x: horner([1, 0, 3, 0,-1,-1],6, x)
f2 = lambda x: 7.0 * np.cos(x-1) - 2.0 * np.sin(x)
f3 = lambda x: 6**x - 2 * x - 5
f4 = lambda x: 2.0 * np.cos(x) - 1.0**x - x**3.0
f5 = lambda x: np.exp(x) - 5.0 * np.cos(x-1)

def wykres(funkc, poczatek, precyzja, rozwiazanie, wynikS, wynikF):
    if poczatek == 0:
        poczatek = -1
    fx = [0.0] * (np.absolute(int(poczatek/precyzja*2.0)))
    fy = [0.0] * (np.absolute(int(poczatek/precyzja*2.0)))
    for i in range (len(fx)):
        fx[i] = poczatek
        fy[i] = funkc(poczatek)
        poczatek = poczatek+precyzja
        
    if rozwiazanie == 1:
        plt.scatter(wynikS, funkc(wynikS), color='darkgoldenrod')
        plt.scatter(wynikF, funkc(wynikF), color='red')
    plt.plot(fx, fy, color='rebeccapurple')
    plt.grid(True)
    
def dostepne_funkcje ():
    print('Dostepne funkcje: ')
    print('[1]  x**5 + 3 * x**3 - x - 1')
    print('[2]  7 * cos(x-1) - 2 * sin(x)')
    print('[3]  6**x - 2 * x - 5')
    print('[4]  2 * cos(x) - 1**x - x**3')
    print('[5]  exp(x) - 5 * cos(x-1)')
    
dostepne_funkcje()
printFunction = input('Wybierz funkcje do narysowania: ')
printFunction = int(printFunction)
if printFunction > 5 and printFunction < 1:
    print("Podano zla wartosc!")
    sys.exit()
    
first = input('Podaj poczatek przedzialu: ')
first = float(first)
if first < 0:
    print('Koniec przedzialu wynosi: ',np.absolute(first))
elif first > 0:
    print('Koniec przedzialu wynosi: ',np.absolute(first*3))
else:
    print('Drobna poprawka, poczatek przedzialu wynosi -1.0\nKoniec przedzialu wynosi: 1.0')
precision = input('Podaj precyzje (proponowane < 1, mniejsza wartosc == wieksza dokladnosc): ')
precision = float(precision)

if printFunction == 1:
    wykres(f1, first, precision, 0, 0, 0)
elif printFunction == 2:
    wykres(f2, first, precision, 0, 0, 0)
elif printFunction == 3:
    wykres(f3, first, precision, 0, 0, 0)
elif printFunction == 4:
    wykres(f4, first, precision, 0, 0, 0)
else:
    wykres(f5, first, precision, 0, 0, 0)
```
